```python
import tensorflow as tf
import sys, os
import matplotlib.pyplot as plt
from PIL import Image, ImageEnhance
import numpy as np
sys.path.append(os.path.abspath(os.path.join('..')))
images_path = "../images/generated_images/"
```

```python
i=0
original_image = tf.keras.utils.load_img(images_path + f"train_batch{0}_sample_{i}_original_image.png")
enhancer_o = ImageEnhance.Brightness(original_image)
original_image = enhancer_o.enhance(8)


predicted_image = tf.keras.utils.load_img(images_path + f"train_batch{0}_sample_{i}_predicted_image.png")
enhancer_p = ImageEnhance.Brightness(predicted_image)
predicted_image = enhancer_p.enhance(8)

ground_truth_image = tf.keras.utils.load_img(images_path + f"train_batch{0}_sample_{i}_ground_truth_image.png")
enhancer_g = ImageEnhance.Brightness(ground_truth_image)
ground_truth_image = enhancer_g.enhance(8)

predicted_image.save("predicted_image_example1.png")
ground_truth_image.save("ground_truth_image_example1.png")
original_image.save("original_image_example1.png") 
fig, ax = plt.subplots(1, 3, figsize=(15, 15))
ax[0].imshow(original_image)
ax[0].set_title("Original image")
ax[1].imshow(predicted_image)
ax[1].set_title("Predicted image")
ax[2].imshow(ground_truth_image)
ax[2].set_title("Ground truth image")
```

```python
nb_batch = 1
nb_sample = 8

fig, axs = plt.subplots(nb_sample, 3, figsize=(20, 20))

for i in range(nb_batch):
    for j in range(nb_sample):
        axs[j, 0].imshow(tf.keras.utils.load_img(images_path + f"train_batch{i}_sample_{j}_original_image.png"), interpolation='nearest')
        axs[j, 1].imshow(tf.keras.utils.load_img(images_path + f"train_batch{i}_sample_{j}_predicted_image.png"), interpolation='nearest')
        axs[j, 2].imshow(tf.keras.utils.load_img(images_path + f"train_batch{i}_sample_{j}_ground_truth_image.png"), interpolation='nearest')
plt.show()
```
